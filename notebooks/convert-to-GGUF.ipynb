## Step 3: Convert to GGUF
This cell will load the base Llama 3.1 8B model, apply your LoRA weights, and export a `q3_k_m` GGUF file. It will take approximately 15-20 minutes.

```python
from unsloth import FastLanguageModel
import os

# 1. Load the Soul (uploaded LoRA checkpoint)
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "/content/my_checkpoint",
    max_seq_length = 2048,
    load_in_4bit = True,
)

# 2. Stitch the soul to the body and export the GGUF
model.save_pretrained_gguf(
    "final_persona_model",
    tokenizer,
    quantization_method = "q3_k_m"
)
```
